# Content Based Recommendation

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
m_cols = ['movie id', 'movie title' ,'release date','video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']

movies = pd.read_csv('data/ml-100k/u.item', sep='|', names=m_cols,
encoding='latin-1')
movies.head()
```

```python
# Remove space in columns
movies.columns = movies.columns.str.replace(" ", "_")
```

```python
movies.head()
```

Seems like video release date has all entries missing

```python
movies.info()
```

```python
movies.movie_title.fillna("", inplace=True)
```

```python
import wordcloud
from wordcloud import WordCloud, STOPWORDS

title_corpus = ' '.join(movies.movie_title)
title_wordcloud = WordCloud(stopwords=STOPWORDS, background_color="black", 
                            width=500, height=300).generate(title_corpus)

plt.figure(figsize=(16, 8))
plt.imshow(title_wordcloud)
plt.axis("off")
```

```python
movies.columns = movies.columns.str.replace("'", "")
```

Visualising the wordcloud for genre

```python
genre_counts = movies.loc[:, "Action":].sum(axis=0)
```

```python
genre_wordcloud = WordCloud(background_color="white")
genre_wordcloud.generate_from_frequencies(genre_counts)

plt.figure(figsize=(16, 8))
plt.imshow(genre_wordcloud)
plt.axis("off");
```
